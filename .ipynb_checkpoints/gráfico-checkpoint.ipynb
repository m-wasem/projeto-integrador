```python
#filtra as faixas etarias
filtradas25a44 = 0
filtradas45a64 = 0
filtradas65a84 = 0
filtradas85 = 0

for item in list:
    df_item = item.data
    filtradas25a44 += df_item['IDADE'][pd.Series((df_item['IDADE'] >= 425) & (df_item['IDADE'] <= 444))].count()

    filtradas45a64 += df_item['IDADE'][pd.Series((df_item['IDADE'] >= 445) & (df_item['IDADE'] <= 464))].count()

    filtradas65a84 += df_item['IDADE'][pd.Series((df_item['IDADE'] >= 465) & (df_item['IDADE'] <= 484))].count()
    
    filtradas85 += df_item['IDADE'][pd.Series(df_item['IDADE'] > 485)].count()

#print(str(filtradas25a44) + " | " + str(filtradas45a64) + " | " + str(filtradas65a84) + " | " + str(filtradas85))
```

```python
#Grafico de faixas etarias

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

legendas = ['25 anos - 44 anos', '45 anos - 64 anos', '65 anos - 84 anos', 'Acima de 80 anos']

data = [filtradas25a44, filtradas45a64, filtradas65a84, filtradas85]

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(legendas[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Faixa etária")

plt.show()
```
